```python
# https://www.statology.org/pandas-check-if-string-contains-multiple-substrings/

import pandas as pd

#create DataFrame
df = pd.DataFrame({'team' : ['Good East Team', 'Good West Team', 'Great East Team',
                             'Great West Team', 'Bad East Team', 'Bad West Team', 'jay'],
                   'points' : [93, 99, 105, 110, 85, 88, 100]})

#view DataFrame
df
```

```python
df['good_or_east'] = df['team'].str.contains('|'.join(['Good', 'East']))
df
```

```python
ar = ['Good', 'jay']
df['good_or_east'] = df['team'].str.contains('|'.join(ar))
df
```

```python
ar = ['Good', 'jay']
df.loc[df['team'].isin(ar)]
```
